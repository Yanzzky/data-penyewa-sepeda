```python
#Menu Cek Data
@app.route("/data_table")
def data_table():
    df = pd.read_csv("/tmp/day_updated.csv")
    return df.to_html(index=False)
```

```python
@app.route("/filter_data")
def filter_data():
    season = request.args.get("season") # Changed from "Musim" to "season"
    df = pd.read_csv("/tmp/day_updated.csv")
    df = df[df["Musim"] == season]
    return df.to_html(index=False)
```

```python
@app.route("/jmlhpenyewa")
def jmlhpenyewa():
    # Membaca file CSV dari lokasi penyimpanan
    df = pd.read_csv("/tmp/day_updated.csv")

    # Melakukan pengelompokan berdasarkan kolom 'Musim'
    # dan menjumlahkan nilai pada kolom 'Total Penyewa'
    df_sum = df.groupby("Musim")["Total Penyewa"].sum().reset_index()

    # Mengembalikan tabel HTML dengan class 'table' agar styling CSS Anda aktif
    return df_sum.to_html(index=False, classes='table')
```

```python
# Menu rata-rata Penyewa
@app.route("/ratapenyewa")
def ratapenyewa():
    # Membaca file CSV
    df = pd.read_csv("/tmp/day_updated.csv")
    df_hasil = df.groupby("Musim")["Total Penyewa"].mean().reset_index()
    df_hasil.columns = ["Kategori Musim", "Total Penyewa"]
    return df_hasil.to_html(index=False, classes='table', border=0)
```

```python
# Menu rata-rata Penyewa
@app.route("/maxpenyewa")
def maxpenyewa():
    # Membaca file CSV
    df = pd.read_csv("/tmp/day_updated.csv")
    df_hasil = df.groupby("Musim")["Total Penyewa"].max().reset_index()
    df_hasil.columns = ["Kategori Musim", "Total Penyewa"]
    return df_hasil.to_html(index=False, classes='table', border=0)
```

```python
#Menu Min Penyewa
@app.route("/minpenyewa")
def minpenyewa():
  df = pd.read_csv("/tmp/day_updated.csv")
  df_hasil = df.groupby("Musim")["Total Penyewa"].min().reset_index()
  df_hasil.columns = ["Kategori Musim", "Total Penyewa"]
  return df_hasil.to_html(index=False, classes='table', border=0)
```
